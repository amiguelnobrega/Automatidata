```python
#payment_type value counts
df.payment_type.value_counts()
#credit card is by far the most common payment type
```

```python
#above code with %, 0.2% disputes seems reasonable
round(df.payment_type.value_counts(1)*100,ndigits=2)
```

```python
#rounded average credit card payment tip
avg_cc_tip = round(df[df['payment_type']==1]['tip_amount'].mean(),ndigits=2)
print('average cc tip = ', avg_cc_tip)
```

```python
#rounded average cash payment tip
avg_cc_tip = round(df[df['payment_type']==2]['tip_amount'].mean(),ndigits=2)
print('average cash tip = ', avg_cc_tip)
```

```python
#df vendorID count
df['VendorID'].value_counts()
```

```python
#above code with %
round(df['VendorID'].value_counts(1)*100,ndigits=2)
```

```python
#mean total amount per vendor
round(df.groupby(['VendorID']).mean(numeric_only=True)[['total_amount']],ndigits=2)
```

```python
#filter df for credit card payments only
credit_card = df[df['payment_type']==1]

#filter credit-card-only data for passenger count
credit_card['passenger_count'].value_counts()

#0 passengers have used credit card 27 times (check possible error)
```

```python
#average tip amount per passenger count for credit card payments
round(credit_card.groupby(['passenger_count']).mean(numeric_only=True)[['tip_amount']], ndigits=2)
```
